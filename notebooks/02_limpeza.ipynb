```python
import pandas as pd
```

```python
df=pd.read_csv("../data/raw/Chocolate Sales.csv")
```

```python
df.head()
```

```python
New_Df=df.copy()
```

```python
New_Df.info()
```

```python
New_Df.head()
```

```python
New_Df['Date']=pd.to_datetime(New_Df['Date'],dayfirst=True)
```

```python
New_Df['Amount']=(
                    New_Df['Amount']
                    .str.replace(',','')
                    .str.replace('$', '')
                    .astype(float)
)
```

```python
New_Df.info()
```

```python
New_Df.head()
```

```python
New_Df.to_csv("../data/processed/Chocolate_Limpo.csv", index=False)
```
